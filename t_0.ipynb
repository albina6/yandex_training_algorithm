# A. Не минимум на отрезке





```python
n, m = map(int, input().split())
list_numb = list(map(int, input().split()))
list_answer = []
for i in range(m):
    a, b = map(int, input().split())
    flag_not_foung = True
    for k in range(a + 1, b + 1):
        if list_numb[k] > list_numb[a]:
            list_answer.append(list_numb[k])
            flag_not_foung = False
            break
        elif list_numb[k] < list_numb[a]:
            list_answer.append(list_numb[a])
            flag_not_foung = False
            break
    if flag_not_foung:
        list_answer.append('NOT FOUND')
print(*list_answer, sep='\n')
```

# C. Путешествие по Москве


```python
import math

def r_alf(x, y):
    alf = math.atan2(y, x)
    r = pow(x * x + y * y, 0.5)
    return r, alf



x_a, y_a, x_b, y_b = map(int, input().split())

r_a, alf_a = r_alf(x_a, y_a)
r_b, alf_b = r_alf(x_b, y_b)

r_def = abs(r_a - r_b)
alf_def = max(alf_a, alf_b) - min(alf_a, alf_b)
length_r = min(r_a, r_b) * alf_def
len_1 = r_a + r_b
len_2 = length_r + r_def

print(min(len_1, len_2))
```

# D. Анаграмма?


```python
def dict_count(char_list):
    dict_c = {}
    for  i in char_list:
        dict_c[i] = dict_c.get(i, 0) + 1 
        
    return dict_c

a = input()
b = input()
a_dict = dict_count(a)
b_dict = dict_count(b)
flag = 'YES'
for key, value in a_dict.items():
    if b_dict.get(key, 0) != value:
        flag = 'NO'
        break
print(flag)
```

# E. Средний уровень


```python
n = int(input())
rating_list = list(map(int, input().split()))

answer = []
sum_list = sum(rating_list)
answer_pred = sum(abs(rating_list[0] - i) for i in rating_list)
for i in range(1, n):
    answer.append(answer_pred)
    k = rating_list[i] - rating_list[i - 1]
    answer_pred = answer_pred + k * (2 * i - n)
answer.append(answer_pred)

print(*answer)
```

```python
n = int(input())
rating_list = list(map(int, input().split()))

answer = []
sum_next = sum(rating_list)
sum_pred = 0

for i in range(0, n):
    elem = rating_list[i]
    sum_next -= elem
    answer_i = (elem * i - sum_pred) + (sum_next - (elem * (n - i - 1)))
    answer.append(answer_i)
    sum_pred += elem
    
print(*answer)
```

# F. Лифт


```python
# k = 2
# n = 3
# s_ = '''\
# 3
# 0
# 1'''
# floor_list = list(map(int, s_.split('\n')))

k = int(input())
n = int(input())
floor_list = list(map(int, (input() for i in range(n))))


sec = 0

index = n - 1
flag = True
while index >= 0 and flag:
    pas_current = floor_list[index]
    sec += 2 * (index + 1) * (pas_current // k)
    if (pas_current % k) > 0:
        sec += 2 * (index + 1) 
        index -= 1
        free_ = k - pas_current % k
        for ind_floor in range(index, -1, -1):
            if floor_list[ind_floor] > 0:
                pas_pred = floor_list[ind_floor]
                
                floor_list[ind_floor] = max(pas_pred - free_, 0)
                free_ = max(free_ - pas_pred, 0)
                
            if index == 0 and free_ >= 0 and floor_list[ind_floor] == 0:
                flag = False
                
            elif free_ == 0:
                index = ind_floor
                break
    else:
        index -= 1
    
print(sec)
```

# G. Кролик учит геометрию


```python
n, m = map(int, input().split())
square = [list(map(int, input().split())) for row in range(n)]
square_new = [[0] * m for i in range(n)]

for column in range(0, m):
    square_new[0][column] = square[0][column]
               
for row in range(0, n):
    square_new[row][0] = square[row][0]
    

max_side = 0
for row in range(1, n):
    for column in range(1, m):
        if square[row][column]:
            k = min(square_new[row - 1][column - 1], square_new[row][column - 1], square_new[row - 1][column]) + 1
            if k > max_side:
                max_side = k
        else:
            k = square[row][column]
        square_new[row][column] = k
        
        
print(max_side)
```

# J. Групповой проект


```python
# str_ ='''\
# 4
# 9 3 5
# 11 7 8
# 28 4 6
# 3 1 2'''
# t, nab_list = str_.split('\n', maxsplit=1)
# t = int(t)
# nab_list = [list(map(int, i.split())) for i in nab_list.split('\n')]

t = int(input())
nab_list = [list(map(int, input().split())) for i in range(t)]

answer = []
for i in nab_list:
    n, a, b = i
    a_count = n // a
    a_free = n % a
    k = b - a
    if a_count * k < a_free:
        answer.append(False)
    else:
        answer.append(True)
print('\n'.join(list(map(lambda x: 'YES' if x else 'NO', answer))))
```
